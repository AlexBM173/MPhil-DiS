# Question 1: Comparing dataset sizes

How big can they possibly be?

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Find some datasets and record sizes

sizes_mb = [11.06, 23550, 807900, 10370, 9, 0.05566, 2.684e11]
labels = ["MNIST", "NASA Exoplanet Archive: Optical spectra", "ImageNet: Raw 1600px images", "Planck Full Mission Frequency Maps",
          "Corona Virus 2019", "World Happiness Report 2021", "Estimated Euclid lifetime raw data"]
```

```python
plt.bar(labels, np.log2(sizes_mb))
plt.xticks(rotation=80, fontsize=5)
plt.ylabel("log2 Size (MB)")
plt.title("Dataset Sizes in log2(MB)")
plt.show()
```
