```python
import numpy as np
import matplotlib.pyplot as plt
import healpy as hp

map_30 = hp.fitsfunc.read_map("LFI_30_GHz_IQU.fits", field=0)
map_30_Q = hp.fitsfunc.read_map("LFI_30_GHz_IQU.fits", field=1)
map_30_U = hp.fitsfunc.read_map("LFI_30_GHz_IQU.fits", field=2)
hp.mollview(map_30, max=2e-4, min=-2e-4)
plt.show()
hp.mollview(map_30_Q, max=2e-4, min=-5e-4)
plt.show()
hp.mollview(map_30_U, max=2e-4, min=-2e-4)
plt.show()
smica_map = hp.fitsfunc.read_map("SMICA_PR2.fits")
hp.mollview(smica_map, max=2e-4, min=-2e-4)
plt.show()
```

```python
ells = np.arange(0, 2048 + 1)
ps_cl = hp.anafast(1e6 * smica_map, lmax=2048)
plt.plot(ells, ps_cl)
plt.show()
best_dl = np.loadtxt("COM_PS.txt")
ps_dl = ells * (ells + 1)/ (2 * np.pi) * ps_cl
plt.plot(ells,ps_dl)
plt.plot(best_dl[:,0], best_dl[:,1])
plt.show()

```
