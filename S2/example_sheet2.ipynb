## Question 2: Plotting the banana

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def banana_2(x, y):
    return 100 * (y - x ** 2) ** 2 + (1 - x) ** 2
```

```python
x = np.arange(-0.5, 3, 0.01)
y = np.arange(-0.5, 3, 0.01)
grid = np.meshgrid(x, y)
levels = np.arange(0, 1, 0.1)
constant = 1
data = np.exp(-banana_2(grid[0], grid[1]) + constant)

plt.contour(x, y, data, levels=levels)
plt.show()
```

## Question 3: Deterministic method

```python
def banana_3(x, y, z):
    return 100 * (y - x ** 2) ** 2 + (1 - x) ** 2 + 100 * (z - y ** 2) ** 2 + (1 - y) ** 2
```

```python
from scipy.integrate import nquad
```

```python
def func_to_integrate(x, y, z):
    return np.exp(-banana_3(x, y, z) + constant) * np.sqrt(x ** 2 + y ** 2 + z ** 2 )

low = -np.inf
upp = np.inf

integral = nquad(func_to_integrate, [[low, upp], [low, upp], [low, upp]])

print(integral)
```
