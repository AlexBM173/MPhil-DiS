## Exercise 1.1 and 1.2

Calculate sin(0.1) using a Taylor series to order 5. Print the value up to 25 decimal places and compute the error.

```python
import numpy as np
import math

def sine_taylor(x, terms):
    result = 0
    for i in range(terms):
        result += (((-1) ** i) * (x ** (2 * i + 1))) / math.factorial((2 * i + 1))
    return result

value = 0.1
order = 5

sin_approx = sine_taylor(value, order)
sin_true = math.sin(value)

print(f"Approximate value of sin({value}) with {order} terms: {sin_approx:.25f}")
print(f"True value of sin({value}): {sin_true:.25f}")
error = sin_approx - sin_true
print(f"Error: {error:.25f}")
```

## Exercise 1.3

Construct a list of all the primes up to a number $N$ using the Sieve of Eratosthenes.

```python
def return_primes(N):
    is_prime = [True] * N
    is_prime[0] = is_prime[1] = False

    for i in range (2, int(N ** 0.5) + 1):
        if is_prime[i]:
            for j in range(i * i, N, i):
                is_prime[j] = False
    primes = [i for i in range(N) if is_prime[i]]
    return primes

print(return_primes(121))

N = 121
print(f"Prime numbers less than {N}: {return_primes(N)}")
```

## Exercise 1.4

Return the first $N$ numbers in Recaman's sequence.

```python
def Recaman(N):
    sequence = [0]
    for i in range(1, N):
        previous = sequence[-1]
        next = previous - i
        if next > 0 and next not in sequence:
            sequence.append(next)
        else:
            sequence.append(next + 2 * i)
    return(sequence)

print(Recaman(100))
```

# Exercise 1.5

Compute a list of numbers that appear in both of two equal length lists.

```python
import random

N = 10
list1 = random.sample(range(1, N * 2), N)
list2 = random.sample(range(1, N * 2), N)

def nums_in_both(list1, list2):
    results = []
    for i in list1:
        if i in list2:
            results.append(i)
    return(results)

print(nums_in_both(list1, list2))

# The previous implementation is not that efficient. We can make it efficient by converting to sets first, which removes duplicates,
# and joining the two sets with the AND operator that creates a set of elements appearing in both sets.

def efficient_nums_in_both(list1, list2):
    return(list(set(list1) & set(list2)))

print(efficient_nums_in_both(list1, list2))
```

## Exercise 1.6

Create a list of all pairs of factors of the number 362,880 as tuples using list comprehensions.

```python
N = 362880

factors = [(x, N // x) for x in range(1, int(N ** 0.5) + 1) if N % x == 0]
print(factors)
```

## Exercise 1.7

Create a simple 1D random walk that is equally likely to move up or down with a fixed step size.

```python
import matplotlib.pyplot as plt

N_steps = 100
starting_val = 0

def random_1D(N_steps, starting_val):
    step_nums = [0]
    pos = [starting_val]
    for i in range(1, N_steps):
        step_nums.append(i)
        step = random.randint(0,1)
        if step == 1:
            pos.append(pos[-1] + 1)
        else:
            pos.append(pos[-1] - 1)
    return((step_nums, pos))


for i in range(5):
    steps, pos = random_1D(N_steps, starting_val)
    plt.plot(steps, pos)
plt.show()
```

## Exercise 2: Geometric Brownian motion

Geometric Brownian motion is a process obeying the stochastic differential equation (SDE):

\begin{equation}
    dY(t) = \mu Y(t)dt + \sigma Y(t) dB(t).
\end{equation}

$B(t)$ is called a Brownian motion, $\mu$ is called the drift, and $\sigma$ is the diffusion coefficient. The solution is:

\begin{equation}
    Y(t) = Y_0 = 0\exp \biggl( \biggl( \mu - \frac{\sigma^2}{2}  \biggr) t + \sigma B(t) \biggr).
