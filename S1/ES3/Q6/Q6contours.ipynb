# Contours and likelihoods

## One-dimensional measurement in a two-dimensional plot

Imagine measuring a parameter $\beta = (22\pm1)^\circ$ that relates to the Cartesian coordinates by

\begin{equation}
    y=(1-x)\tan\beta.
\end{equation}

Draw contours containing $68.3\%$ and $95.4\%$ of the probability distribution for $\beta$ but on a plot of $x$ versus $y$.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns

beta = 22
beta_sigma = 1
def y(x, beta):
    return(1 - x) * np.tan(beta)

x = np.linspace(0, 1, 100)
y_values = y(x, beta)


y_lower = y(x, beta - beta_sigma)
y_upper = y(x, beta + beta_sigma)

y_double_lower = y(x, beta - 2 * beta_sigma)
y_double_upper = y(x, beta + 2 * beta_sigma)

plt.figure(figsize=(10, 6))
plt.plot(x, y_values, label='y(x, beta)', color='blue')
plt.fill_between(x, y_lower, y_upper, color='blue', alpha=0.3, label='Uncertainty Range (±1σ)')
plt.fill_between(x, y_double_lower, y_double_upper, color='blue', alpha=0.2, label='Uncertainty Range (±2σ)')
plt.title('Plot of y(x, beta) with uncertainty')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.ylim(-2.5, 2.5)
plt.xlim(0, 1)
plt.show()
```
