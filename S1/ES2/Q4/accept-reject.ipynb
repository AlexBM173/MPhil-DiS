# Accept-reject sampler

Find the maximum of the function you wish to sample from in a given range $f_\text{max}$. Generate an $x$ value in the range uniformly then generate a value between 0 and $f_\text{max}$ uniformly. If $f(x)$ is greater than this value, keep the $x$ value.

Start with

\begin{equation}
    f(x) = \cos^2(x)
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

lower_lim = 0
upper_lim = 10

def distribution_function(x):
    f = np.cos(x) ** 2
    return f

f_max = 1

x_vals = np.random.uniform(lower_lim, upper_lim, 10000)
x_test_vals = np.random.uniform(0, f_max, 10000)
samples = [i for i in x_vals if x_test_vals[np.where(x_vals == i)[0][0]] < distribution_function(i)]
sample_vals = [i for i in x_test_vals if i < distribution_function(x_vals[np.where(x_test_vals == i)[0][0]])]
rejects = [i for i in x_vals if x_test_vals[np.where(x_vals == i)[0][0]] >= distribution_function(i)]
reject_vals = [i for i in x_test_vals if i >= distribution_function(x_vals[np.where(x_test_vals == i)[0][0]])]
print(samples)

x = np.linspace(lower_lim, upper_lim, 1000)
y = distribution_function(x)
plt.plot(x, y, label='Distribution Function', color='blue')
plt.scatter(samples, sample_vals, color='green', s=1, label='Accepted Samples')
plt.scatter(rejects, reject_vals, color='red', s=1, label='Rejected Samples')
plt.title('Rejection Sampling')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

Next function:

\begin{equation}
    f(x)=\frac{\sin(x)+\cos(x)}{\sinh (x)+\cosh(x)} + 25.
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

lower_lim = 0
upper_lim = 10

def distribution_function(x):
    f = (np.sin(x)+ np.cos(x)) / (np.sinh(x) + np.cosh(x)) + 25
    return f

f_max = 26

x_vals = np.random.uniform(lower_lim, upper_lim, 10000)
x_test_vals = np.random.uniform(0, f_max, 10000)
samples = [i for i in x_vals if x_test_vals[np.where(x_vals == i)[0][0]] < distribution_function(i)]
sample_vals = [i for i in x_test_vals if i < distribution_function(x_vals[np.where(x_test_vals == i)[0][0]])]
rejects = [i for i in x_vals if x_test_vals[np.where(x_vals == i)[0][0]] >= distribution_function(i)]
reject_vals = [i for i in x_test_vals if i >= distribution_function(x_vals[np.where(x_test_vals == i)[0][0]])]
print(samples)

x = np.linspace(lower_lim, upper_lim, 1000)
y = distribution_function(x)
plt.plot(x, y, label='Distribution Function', color='blue')
plt.scatter(samples, sample_vals, color='green', s=1, label='Accepted Samples')
plt.scatter(rejects, reject_vals, color='red', s=1, label='Rejected Samples')
plt.title('Rejection Sampling')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

Try with a standard normal distribution.

```python
from scipy.stats import norm

lower_lim = -1
upper_lim = 1
def distribution_function(x):
    f = norm.pdf(x, loc=0, scale=1)
    return f

f_max = 1/np.sqrt(2 * np.pi)

x_vals = np.random.uniform(lower_lim, upper_lim, 10000)
x_test_vals = np.random.uniform(0, f_max, 10000)
samples = [i for i in x_vals if x_test_vals[np.where(x_vals == i)[0][0]] < distribution_function(i)]
sample_vals = [i for i in x_test_vals if i < distribution_function(x_vals[np.where(x_test_vals == i)[0][0]])]
rejects = [i for i in x_vals if x_test_vals[np.where(x_vals == i)[0][0]] >= distribution_function(i)]
reject_vals = [i for i in x_test_vals if i >= distribution_function(x_vals[np.where(x_test_vals == i)[0][0]])]
print(samples)

x = np.linspace(lower_lim, upper_lim, 1000)
y = distribution_function(x)
plt.plot(x, y, label='Distribution Function', color='blue')
plt.scatter(samples, sample_vals, color='green', s=1, label='Accepted Samples')
plt.scatter(rejects, reject_vals, color='red', s=1, label='Rejected Samples')
plt.title('Rejection Sampling')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
efficiency = []
sample_ranges = []

for i in range(1, 21):
    lower_lim = -i
    upper_lim = i
    f_max = 1/np.sqrt(2 * np.pi)

    x_vals = np.random.uniform(lower_lim, upper_lim, 10000)
    x_test_vals = np.random.uniform(0, f_max, 10000)
    samples = [i for i in x_vals if x_test_vals[np.where(x_vals == i)[0][0]] < distribution_function(i)]
    efficiency_value = len(samples) / 10000
    sample_ranges.append(i)
    efficiency.append(efficiency_value)
    print(f'Sample Range: {i}, Efficiency: {efficiency_value}')
    
plt.plot(sample_ranges, efficiency)
plt.title('Efficiency of Rejection Sampling vs Sample Range')
plt.xlabel('Sample Range ()')
plt.ylabel('Efficiency')
plt.show()
```
